Generally  , when performing matrix multiplication on two tensors and one of the axes doesn't line up , you will transpose (rather than reshape) one of the tensors to get satisfy the matrix multiplication rules

Aggregating tensors = considering them from multiple values down to a smaller amount of values

Aggregating a tensor means reducing it into a smaller shape or single value by applying operations like:

Sum

Mean

Max / Min

Prod (product)

```python
# create a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0 ,100 , size=50))
E
```

```python
tf.size(E) , E.shape , E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the sum of the tensor
tf.reduce_sum(E)
```

```python
# variance
F= tf.cast(E, dtype=tf.float16)
variance = tf.math.reduce_variance(F)
variance.numpy()
```

```python
# Standard Deviation
st_dev = tf.math.reduce_std(F)
st_dev.numpy()
```

```python
import tensorflow_probability as tfp
# Find the Variance
var = tfp.stats.variance(E)
var.numpy()
```

```python
# find the Stand Deviation
std = tfp.stats.stddev(F)   # requires the tensors only in datatype of  float
std.numpy()
```

```python
# Minimum
tf.minimum(X,y)
```

```python
# creating a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape =[50])
F
```

```python
# find the positional maximum
tf.argmax(F)
```

```python
# index on our largest value position
F[tf.argmax(F)]
```

```python
# find the max value of F
tf.reduce_max(F)
```

```python
# chack for equality
assert F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# find the positinal minimum
tf.argmin(F)
```

```python
# find the minimum using the positional minimum index
F[tf.argmin(F)]
```
